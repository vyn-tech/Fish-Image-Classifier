#### Resize image to 128x128
This small size will lead to fast training, lower memory use. However it might miss fine grain features.
Can be modified.

```python
# import necessary packages/libraries
# !pip install opencv-python
import cv2
import os
from PIL import Image
```

```python
# set working directory
project_folder = r'C:\Users\Vy\Documents\Lewisville_Lake_Fish _Classification'

# fish folders 
base = os.path.join(project_folder, 'Data')
```

```python
# resize images
def resize_images(folder_path, size=(128, 128)):
    for fish_species in os.listdir(folder_path): # fish folders
        species_path = os.path.join(folder_path, fish_species)
        
        for img_file in os.listdir(species_path): # each image file species folder
            img_path = os.path.join(species_path, img_file)
            # read in image 
            img = cv2.imread(img_path) 

            # resize
            if img is not None:
                resized_img = cv2.resize(img, size)
                # write back to folder
                cv2.imwrite(img_path, resized_img)                
```

```python
# apply function
resize_images(base)
```
